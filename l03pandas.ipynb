```python
df['area'] = df['Площадь, тыс.км2']
```

```python
df.drop([24],inplace=True)
```

```python
def replace_comma(s):
    return s.replace(',','.') if type(s)==str else '0'
```

```python
df['area'] = df['area'].apply(replace_comma)
```

```python
df['area'] = df['area'].astype(float)
```

```python
df['Плотность населения'] = df['Численность населения\xa0(2002\xa0 г.), чел']/(1000*df['area'])
```

```python
df['Площадь, тыс.км2']= df['area']
```
